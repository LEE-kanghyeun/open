```python
import pandas as pd 

kuj_list = kuj.split("\n")

area_list = []
addr_list = []

for i in range(len(kuj_list)):
  if i%3 == 0:
    area_list.append(float(kuj_list[i].split(" ")[2].replace(',', '')))
  if i%3 == 1:
    addr_list.append(str(kuj_list[i]))


kuj_df = pd.DataFrame({'addr': addr_list, 'area': area_list} )

kuj_df['type'] = '국유지'
kuj_df
```

```python
# from geopy.geocoders import Nominatim
# https://parkgihyeon.github.io/project/geocoding-api/ ,  좌표와 주소간 변환

from geopy.geocoders import Nominatim

def geocoding(address):
  try:
    geolocoder = Nominatim(user_agent = 'South Korea', timeout=None)
    geo = geolocoder.geocode(address)
    crd = {"lat": str(geo.latitude), "long": str(geo.longitude)}

    return crd
  except:
    crd = {"lat": None, "long": None}

    return crd

x_list = []
y_list = []

for i in kuj_df['addr']:
  # print(i)
  crd = geocoding(i)
  x_list.append(crd['long'])
  y_list.append(crd['lat'])

kuj_df['x'] = x_list
kuj_df['y'] = y_list

kuj_df
```

```python
# https://velog.io/@choonsik_mom/Naver-API%EB%A1%9C-%EA%B8%B8%EC%B0%BE%EA%B8%B0with-python
import json
import urllib
from urllib.request import Request, urlopen
# *-- 3개의 주소 geocoding으로 변환한다.(출발지, 도착지, 경유지) --*
start = '서울특별시 종로구 청운동 108-22'
goal = '서울특별시 강동구 상일동 산 33'
waypoint = '경기도 수원시 장안구 서부로 2149'

# 주소에 geocoding 적용하는 함수를 작성.
def get_location(loc) :
  client_id = 'cgg8claahp'
  client_secret = '비밀번호'
  url = f"https://naveropenapi.apigw.ntruss.com/map-geocode/v2/geocode?query=" \
        + urllib.parse.quote(loc)
  
  # 주소 변환
  request = urllib.request.Request(url)
  request.add_header('X-NCP-APIGW-API-KEY-ID', client_id)
  request.add_header('X-NCP-APIGW-API-KEY', client_secret)
  
  response = urlopen(request)
  res = response.getcode()
  
  if (res == 200) : # 응답이 정상적으로 완료되면 200을 return한다
      response_body = response.read().decode('utf-8')
      response_body = json.loads(response_body)
      # print(response_body)
      # 주소가 존재할 경우 total count == 1이 반환됨.
      if response_body['meta']['totalCount'] == 1 : 
        # 위도, 경도 좌표를 받아와서 return해 줌.
          lat = response_body['addresses'][0]['y']
          lng = response_body['addresses'][0]['x']
          return (lng, lat)
      else :
          print('location not exist')
          lat = None
          lng = None
          return (lng, lat)
      
  else :
      print('ERROR')

        
#  함수 적용
start = get_location(start)
goal = get_location(goal)
waypoint = get_location(waypoint)
```

```python
## 동추출을 위해 지번 주소로 모두 변경

# https://velog.io/@choonsik_mom/Naver-API%EB%A1%9C-%EA%B8%B8%EC%B0%BE%EA%B8%B0with-python
import json
import urllib
from urllib.request import Request, urlopen
# *-- 3개의 주소 geocoding으로 변환한다.(출발지, 도착지, 경유지) --*
start = '서울 강남구 일원로 81 삼성의료원'
goal = '서울특별시 강남구 율현동 밤고개로24길(율현동 56)'
waypoint = '경기도 수원시 장안구 서부로 2149'

# 주소에 geocoding 적용하는 함수를 작성.
def get_jaddress(loc) :
  client_id = 'cgg8claahp'
  client_secret = 'uYx6Ebvrt9gbaP2ls4e0s9QAyXGUHj19CPrZMdif'
  url = f"https://naveropenapi.apigw.ntruss.com/map-geocode/v2/geocode?query=" \
        + urllib.parse.quote(loc)
  
  # 주소 변환
  request = urllib.request.Request(url)
  request.add_header('X-NCP-APIGW-API-KEY-ID', client_id)
  request.add_header('X-NCP-APIGW-API-KEY', client_secret)
  
  response = urlopen(request)
  res = response.getcode()
  
  if (res == 200) : # 응답이 정상적으로 완료되면 200을 return한다
      response_body = response.read().decode('utf-8')
      response_body = json.loads(response_body)
      # print(response_body)
      # 주소가 존재할 경우 total count == 1이 반환됨.
      if response_body['meta']['totalCount'] == 1 : 
        # 위도, 경도 좌표를 받아와서 return해 줌.
          jibunaddr = response_body['addresses'][0]['jibunAddress']
          return jibunaddr
      else :
          print('location not exist')
          jibunaddr = 'None'

          return jibunaddr
      
  else :
      print('ERROR')

        
#  함수 적용
start = get_jaddress(start)
goal = get_jaddress(goal)
waypoint = get_jaddress(waypoint)

start, goal, waypoint
```
